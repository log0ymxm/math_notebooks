```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

%pylab inline
```

```python
x = np.linspace(-10,10,100)
v = (stats.norm.pdf(x) / stats.norm.cdf(x))
w = v * (v + x)

plt.subplot(2,2,1)
plt.plot(x, stats.t.pdf(x, 1))
plt.title('PDF')

plt.subplot(2,2,2)
plt.plot(x, stats.t.cdf(x, 1))
plt.title('CDF')

plt.subplot(2,2,3)
plt.plot(x, v);
plt.title('V')

plt.subplot(2,2,4)
plt.plot(x, w)
plt.title('W')
```

```python
df = 3
x = np.linspace(-10,10,100)
v = (stats.t.pdf(x, df) / stats.t.cdf(x, df))
w = v * (v + x)

plt.subplot(2,2,1)
plt.plot(x, stats.t.pdf(x, 1))
plt.title('PDF')

plt.subplot(2,2,2)
plt.plot(x, stats.t.cdf(x, 1))
plt.title('CDF')

plt.subplot(2,2,3)
plt.plot(x, v);
plt.title('V')

plt.subplot(2,2,4)
plt.plot(x, w)
plt.title('W')
```
